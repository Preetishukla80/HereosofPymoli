```python
# Dependencies and Setup
import pandas as pd
import numpy as np
```

```python
# Raw data file
file = "Resources/purchase_data.csv"
```

```python
# Read purchasing file and store into pandas data frame
purchase_data = pd.read_csv(file) 
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
player_count = len(purchase_data["SN"].value_counts()) 
pd.DataFrame([player_count], columns = ["Total Players"])
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Unique Items
unique_items = len(purchase_data["Item ID"].value_counts())

# Average Price
average_price = round(purchase_data["Price"].mean(), 2)  

# Number of purchases
number_of_purchases = (purchase_data["Price"].count()) 

# Total money spent
total_revenue = round(purchase_data["Price"].sum(), 2) 

# List for analysis
purchase_analysis = []

# Appending values to list
purchase_analysis.append(unique_items)
purchase_analysis.append("$" + str(average_price))
purchase_analysis.append(number_of_purchases)
purchase_analysis.append("$" + str(total_revenue))

# Converting to a DataFrame
pd.DataFrame([purchase_analysis], columns = ["Number of Unique Items", "Average Price", "Number of Purchases", "Total Revenue"])
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#Grouping by Gender and Counting 

Total_gender = purchase_data["Gender"].count()
Male_gender = purchase_data["Gender"].value_counts()['Male']
Female_gender = purchase_data["Gender"].value_counts()['Female']
Non_Disclosed = Total_gender - Male_gender - Female_gender

Male_Percent = (Male_gender / Total_gender) * 100
Female_Percent = (Female_gender / Total_gender) * 100
Non_Disclosed_Percent = (Non_Disclosed / Total_gender) * 100

# Create new DataFrame
Gender_data = pd.DataFrame({"": ['Male', 'Female', 'Other/Non-Disclosed'],
                            "Percentage of Players": [Male_Percent, Female_Percent, Non_Disclosed_Percent],
                            "Total Count": [Male_gender, Female_gender, Non_Disclosed]})

# DataFrame formatting
Gender_data["Percentage of Players"] = Gender_data["Percentage of Players"].map("{:.2f}%".format)
Gender_data = Gender_data.set_index('')
Gender_data
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
gender_data = purchase_data[["SN","Gender","Price"]]
gender_counts = gender_data["Gender"].value_counts()

# Purchase counts
purchase_counts = [gender_counts[0],gender_counts[1],gender_counts[2]]

gender_data = gender_data.groupby("Gender")
gender_total_spent = gender_data.sum()
gender_total_spent

# Total Purchase Value
total_purchase_value_gender = [gender_total_spent.iloc[1,0], gender_total_spent.iloc[0,0], gender_total_spent.iloc[2,0]]

# Average Purchase Price
avg_purchase_gender = [gender_total_spent.iloc[1,0]/gender_counts[0], gender_total_spent.iloc[0,0]/gender_counts[1], gender_total_spent.iloc[2,0]/gender_counts[2]]

# Normalized Totals
norm_total_gender = [gender_total_spent.iloc[1,0]/gender_counts[0], gender_total_spent.iloc[0,0]/gender_counts[1], gender_total_spent.iloc[2,0]/gender_counts[2]]

# Creating DataFrame & setting index
purchase_analysis_g_df = pd.DataFrame({
    "Purchase Count": purchase_counts,
    "Average Purchase Price": avg_purchase_gender,
    "Total Purchase Value": total_purchase_value_gender,
    "Normalized Totals": norm_total_gender,
    "Gender": ["Male", "Female", "Other / Non-Disclosed"]
})
purchase_analysis_g_df = purchase_analysis_g_df.set_index("Gender")
purchase_analysis_g_df = purchase_analysis_g_df[["Purchase Count", "Average Purchase Price", "Total Purchase Value",
                                                 "Normalized Totals"]]

# Formatting Prices
purchase_analysis_g_df.style.format({"Average Purchase Price": "${:.2f}", "Normalized Totals": "${:.2f}",
                                     "Total Purchase Value":"${:.2f}"})
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Establish bins for ages
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

Heroes_data = purchase_data.loc[:,["SN","Age"]]
Heroes_data["Age Ranges"] = pd.cut(Heroes_data["Age"], age_bins, labels=group_names)
Heroes_data

age_demographics_totals = Heroes_data["Age Ranges"].value_counts()
age_demographics_percents = (age_demographics_totals / 780)*100
age_demographics = pd.DataFrame({"Percentage of Total": age_demographics_percents, "Total Counts": age_demographics_totals})

age_demographics.sort_index()
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Establish bins for ages
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

Hereos_data = purchase_data.copy()
Hereos_data["Age Ranges"]=pd.cut(Hereos_data["Age"],age_bins,labels = group_names)
age_demographics = Hereos_data.groupby(['Age Ranges'])

#Print(Grouped_Hereos_data)
#Grouped_Hereos_data.count()

purchase_count= age_demographics["Age"].count()
average_price= age_demographics["Price"].mean()
average_total= age_demographics["Price"].sum()
average_total_pp=average_total/purchase_count

summary_age_analysis= pd.DataFrame({"Purchase count":purchase_count,"Average Purchase Price":average_price,
                                    "Total Purchase Value":average_total,"Average Total Per Person":average_total_pp})
summary_age_analysis
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
top_spenders_df = purchase_data['Item ID'].groupby(purchase_data['SN']).count()

# Creating the initial column in the Dataframe
top_spenders_df = pd.DataFrame(data=top_spenders_df)
top_spenders_df.columns = ['Purchase Count']

# Calculating the column values
top_spenders_df['Average Purchase Price'] = round(purchase_data['Price'].groupby(purchase_data['SN']).mean(),2)
top_spenders_df['Total Purchase Value'] = purchase_data['Price'].groupby(purchase_data['SN']).sum()

top_spenders_df.sort_values(by=['Total Purchase Value'], ascending=False, inplace=True)

# Formatting the currency columns into $0.00 for ease of viewing
top_spenders_df['Average Purchase Price'] = top_spenders_df['Average Purchase Price'].map('${:,.2f}'.format)
top_spenders_df['Total Purchase Value'] = top_spenders_df['Total Purchase Value'].map('${:,.2f}'.format)

top_spenders_df.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Getting Purchase Count and Total Purchase Value
most_items_df = purchase_data.groupby(['Item ID', 'Item Name','Price'])['Price'].agg(['count','sum'])
most_items_df.columns = ['Purchase Count', 'Total Purchase Value']

# Reformatting the index
most_items_df.reset_index(inplace=True)
most_items_df.set_index(['Item ID','Item Name'] ,inplace=True)

# Renaming Dataframe columns
most_items_df = most_items_df[['Purchase Count', 'Price', 'Total Purchase Value']]

# Sorting by most popular (Highest Purchase Count)
most_popular_items_df = most_items_df.sort_values(by='Purchase Count', ascending=False)

# Formatting the currency columns into $0.00 for ease of viewing
most_popular_items_df['Price'] = most_popular_items_df['Price'].map('${:,.2f}'.format)
most_popular_items_df['Total Purchase Value'] = most_popular_items_df['Total Purchase Value'].map('${:,.2f}'.format)

most_popular_items_df.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
# Sorting by most profitable (Total Purchase Value)
most_profitable_items_df = most_items_df.sort_values(by='Total Purchase Value', ascending=False)

# Formatting the currency columns into $0.00 for ease of viewing
most_profitable_items_df['Price'] = most_profitable_items_df['Price'].map('${:,.2f}'.format)
most_profitable_items_df['Total Purchase Value'] = most_profitable_items_df['Total Purchase Value'].map('${:,.2f}'.format)

most_profitable_items_df.head()
```
